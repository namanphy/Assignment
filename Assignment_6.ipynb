```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv"
ds = pd.read_csv(url)
ds.head(10)
```

```python
ds.tail(10)
```

```python
ds.shape
```

```python
#Univariate plot for minimum temperature
MinTemp=ds['MinTemp']
sns.distplot(MinTemp, kde=True, rug=True)
plt.show()
plt.close()
```

```python
#Univariate plot for Maximum temperature
MaxTemp=ds['MaxTemp']
sns.distplot(MaxTemp, kde=True, rug=True)
plt.show()
plt.close()
```

```python
#KDE plot comparison
for col in ['MinTemp', 'MaxTemp']:
    sns.kdeplot(ds[col], shade=True)
plt.show()
plt.close()
```

```python
#scatter plot with regression line
sns.lmplot(data = ds , x = 'MaxTemp' , y = 'MinTemp')
plt.show()
plt.close()
```

```python
#KDE plot
sns.kdeplot(ds['Rainfall'], shade=True, bw=2, color='y' )
plt.show()
plt.close()
```

```python
#Jointplot Rainfall Vs. Evaporation
sns.jointplot(x='Rainfall', y='Evaporation', data=ds, size=12, ratio=15, space=0.5)
plt.show()
plt.close()
```

```python
#scatterplot with histogram and r
sns.jointplot(data = ds , x = 'Sunshine' , y = 'Evaporation' , color = 'r')
plt.show()
plt.close()
```

```python
#resid with r
sns.jointplot(data = ds , x = 'Temp9am' , y = 'Temp3pm' , color = 'black' , kind = 'resid')
plt.show()
plt.close()
```

```python
#strip plot = categorical data vs. numerical data graph
sns.stripplot(data = ds , x = 'WindDir9am' , y = 'WindSpeed9am' , color = 'grey')
plt.show()
plt.close()
```

```python
#strip plot= categorical data vs. numerical data graph
sns.stripplot(data = ds , x = 'WindDir3pm' , y = 'WindSpeed3pm' , color = 'orange')
plt.show()
plt.close()
```

```python
#swarm plot = categorical data vs. numerical data graph without overlapping
sns.swarmplot(data = ds , x = 'Location' , y = 'Pressure9am' , color = 'grey')
plt.show()
plt.close()
```

```python
#swarm plot 
sns.swarmplot(data = ds , x = 'RainToday' , y = 'Humidity3pm' , color = 'indigo')
plt.show()
plt.close()
```

```python
#histogram and rug
sns.distplot(ds['Pressure3pm'] , hist = True , rug = True , kde = False)
plt.show()
plt.close()
```

```python
#kde and rug
sns.distplot(ds['Pressure9am'] , hist = False , rug = True , kde = True , color = 'r')
plt.show()
plt.close()
```

```python
#hex with r
sns.jointplot(data = ds , x = 'MaxTemp' , y = 'Humidity9am' , color = 'y' , kind = 'hex')
plt.show()
plt.close()
```

```python
#Converting Wind Directions to appropiate radian angle values

from math import radians
from collections import Counter

dir={    'ENE': 22.5,
         'SSE': 292.5,
         'SE': 315,
         'E': 0,
         'S': 270,
         'N': 90,
         'WNW': 167.5,
         'ESE': 337.5,
         'NE': 45,
         'NNE': 77.5,
         'NNW': 112.5,
         'SW': 225,
         'W': 180,
         'WSW': 202.5,
         'SSW': 247.5,
         'NW': 135
    }

a=0
for x in ds['WindGustDir']:
    #print(x)
    if x in dir:
        print("direction =" + x)
        ds.ix[a,'WindGustDir'] = dir[x]
        a=a+1
        #print(x)
    else:
         x = None

ds['WindGustDir'].drop(ds.index[363:366])
ds=ds.convert_objects(convert_numeric=True)
ds.dtypes
```

```python
#Plot showing wind gust speed to directions

ax = plt.subplot(111, polar=True)
ax.scatter(x=[radians(x) for x in ds['WindGustDir'].values], y=ds['WindGustSpeed'].values)
ax.set_theta_zero_location('E')
#(lines,labels) = thetagrids( range(0,360,18), ("XXXXX","YYYYY","ZZZZ","PPPP"))
ax.set_theta_direction(1)
plt.show()
```
