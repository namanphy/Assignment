```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url1=r'C:\Users\naman\Downloads\Compressed\Bike-Sharing-Dataset\hour.csv'
df=pd.read_csv(url1)
```

```python
#Question 1
df.head()
```

The given dataset shows the data about bike rental services by BikeRental Company in USA. It depicts how can environmental and seasonal settings. For instance, weather conditions, precipitation, day of week, season, hour of the day etc. can affect the rental behaviors.

```python
#Removing unnecessary columns from the data
df=df.drop(['instant','dteday'], axis=1)
df.head()
```

-------------------------------------------------

```python
df['weathersit'].value_counts()
```

```python
sns.countplot(df['weathersit'])
plt.show()
plt.close()
```

This shows that on a clear day the number of bikes hired are more then on a rough weather.

---------------------------------------------------

```python
df['workingday'].value_counts()
```

```python
sns.countplot(df['workingday'])
plt.show()
plt.close()
```

This shows that most people prefer hiring bikes on working days.

---------------------------------

```python
df['hr'].value_counts()
```

```python
sns.kdeplot(df['hr'], shade=True, color="y")
plt.show()
plt.close()
```

From the above density plot, it is visible that most common hours for bike rental are from 3 to 22 hours.

```python
sns.countplot(x='weekday',hue='workingday',data=df)
plt.show()
plt.close()
```

```python
corr_matrix=df[['temp','casual','registered']].corr(method='pearson')
sns.heatmap(corr_matrix)
plt.show()
plt.close()
```

```python
sns.jointplot(x = 'hum', y = 'cnt', data = df, kind= 'kde')
plt.show()
plt.close()
```

```python
sns.lmplot('hum', 'cnt', data = df, hue = 'workingday', fit_reg = False)
plt.show()
plt.close()
```

```python
sns.pairplot(df[['season', 'mnth', 'holiday', 'weekday',
       'workingday', 'weathersit', 'cnt']])
plt.show()
plt.close()
```
