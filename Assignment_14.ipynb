```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
df=datasets.load_digits()
```

```python
plt.figure(figsize = (20,4))
for index , (image,label) in enumerate(zip(df.data[0:5] , df.target[0:5])):
  plt.subplot(1,5,index+1)
  plt.imshow(np.reshape(image , (8,8)) , cmap = plt.cm.gray)
```

```python
x = pd.DataFrame(np.array(df.data))
y = pd.DataFrame(df.target)
print("shape of X : ", x.shape)
print("shape of Y : ",y.shape)
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit_transform(x)
```

```python
x=pd.DataFrame(x)
```

```python
x.head()
```

```python
from sklearn.decomposition import PCA
model=PCA(n_components=22)
principalcomp=model.fit_transform(x)

final_x=pd.DataFrame(principalcomp)
final_x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(final_x, y, test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,predict)
score          #The accuracy of the model is 96.44%
```

```python
print('The accuracy is ' + str(100*score))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict))
```
